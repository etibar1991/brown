# Шаг 5. Проверка гипотез

**Xbox One и PC**

Проверим гипотезу про равенство средних пользовательских оценок двух платформ. Для это используем Т-тест. Нулевую гипотезу строим только из равенства/неравенства.

**Нулевая гипотеза** - средние пользовательские рейтинги равны.

**Альтернативная гипотеза** - средние пользовательские рейтинги различаются.

Значение альфа - 5%.

```python
alpha = 0.05  
results = st.ttest_ind(games_final[(games_final.platform == "XOne") & (games_final.user_score.notna())]['user_score'], games_final[(games_final.platform == "PC") & (games_final.user_score.notna())]['user_score']
)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
 
```

```python
print(games_final.query('platform in ["XOne", "PC"]').pivot_table(index='platform', values='user_score', aggfunc=['mean', 'median']))
# проверим цифры
```

Можно предположить, что средние оценки на двух платформах не отличаются.

Проверим гипотезу о неравенстве средних пользовательских оценок жанров экшен и спортивные симуляторы. Используем для этого Т-тест. Нулевую гипотезу строим только из равенства/неравенства.

**Нулевая гипотеза** - средние пользовательские рейтинги двух жанров равны.

**Альтернативная гипотеза** - средние пользовательские рейтинги жанров различаются.

Значение альфа 5%.

```python
alpha = 0.05 
results = st.ttest_ind(games_final[(games_final.genre == "action") & (games_final.user_score.notna())]['user_score'], games_final[(games_final.genre == "sports") & (games_final.user_score.notna())]['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
    
```
