* –†–∞—Å—Å–º–æ—Ç—Ä–µ—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ - –ø–æ —Ç–∞–±–ª–∏—Ü–µ —Å –∑–∞–∫–∞–∑–∞–º–∏;
* –ü–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–µ –ø–æ–ø–∞–¥–∞—é—Ç –ª–∏ –∫–∞–∫–∏–µ-—Ç–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ –≤ –æ–±–µ –≥—Ä—É–ø–ø—ã - –ø–æ —Ç–∞–±–ª–∏—Ü–µ —Å –∑–∞–∫–∞–∑–∞–º–∏.
* –ü–æ—Å–º–æ—Ç—Ä–µ—Ç—å –¥–∏–Ω–∞–º–∏–∫—É –ø–æ—Å–µ—Ç–∏—Ç–µ–ª–µ–π –ø–æ –¥–Ω—è–º –ø–æ –≥—Ä—É–ø–ø–∞–º (visitos).

**2.1 –ö—É–º–º—É–ª—è—Ç–∏–≤–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º**

–ù–µ–æ–±—Ö–æ–¥–∏–º–æ —Å–æ–∑–¥–∞—Ç—å –µ–¥–∏–Ω—É—é —Ç–∞–±–ª–∏—Ü—É, —Å–æ–¥–µ—Ä–∂–∞—â—É—é –¥–∞–Ω–Ω—ã–µ –æ –¥–∞—Ç–∞—Ö, –≥—Ä—É–ø–ø–∞—Ö –∏ –∫—É–º—É–ª—è—Ç–∏–≤–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö.

```python
grouped_1 = orders[['date', 'group']].drop_duplicates()

grouped_1 = grouped_1.apply(
    lambda x: orders[
        np.logical_and(
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transaction_id': 'nunique',
            'visitor_id': 'nunique',
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

grouped_1.head(10)
```

–¢–µ–ø–µ—Ä—å –ø–æ–ª—É—á–∏–º –∞–≥—Ä–µ–≥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –∫—É–º–º—É–ª—è—Ç–∏–≤–∞–º–∏ —Å —Ç–∞–±–ª–∏—Ü–µ —Å –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è–º–∏.

```python
grouped_2 = grouped_1.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg(
        {
            'date': 'max', 
            'group': 'max', 
            'visitors': 'sum'
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

grouped_2.head(10)
```

```python
cummulative_table = grouped_1.merge(grouped_2, 
                                    left_on = ['date', 'group'], 
                                    right_on = ['date', 'group'])
cummulative_table.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']
#–æ–±—ä–µ–¥–∏–Ω–∏–º –≤—Å–µ –≤ –æ–¥–Ω—É —Ç–∞–±–ª–∏—Ü—É
```

```python
cummulative_table['conversion'] = cummulative_table['orders'] / cummulative_table['visitors']
#–ø–æ—Å—á–∏—Ç–∞–µ–º –∫–æ–≤–µ—Ä—Å–∏—é
cummulative_table['av_check'] = round(cummulative_table['revenue'] / cummulative_table['orders'],0)
# –∏ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫
cummulative_table.head()
```

–¢–µ–ø–µ—Ä—å –ø–æ—Å—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫ –∫—É–º–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º.

```python
cum_rev_a = cummulative_table[cummulative_table['group'] == 'A'][['date', 'revenue', 'orders']]
cum_rev_b = cummulative_table[cummulative_table['group'] == 'B'][['date', 'revenue', 'orders']]

plt.plot(cum_rev_a['date'], cum_rev_a['revenue'], label = 'A')
plt.plot(cum_rev_b['date'], cum_rev_b['revenue'], label = 'B')
plt.ylabel('–í—ã—Ä—É—á–∫–∞, –º–ª–Ω.—Ä—É–±–ª–µ–π')
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ –∫—É–º–º—É–ª—è—Ç–∏–≤–Ω–æ–π –≤—ã—Ä—É—á–∫–∏ –ø–æ –≥—Ä—É–ø–ø–∞–º', color = 'Blue')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

–°–µ–π—á–∞—Å –ø—Ä–∏ –∞–Ω–∞–ª–∏–∑–µ –≥—Ä–∞—Ñ–∏–∫–∞ –º—ã –Ω–µ —É—á–∏—Ç—ã–≤–∞–µ–º, —á—Ç–æ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤–∏–¥–µ–Ω —Ä–æ—Å—Ç –≤—ã—Ä—É—á–∫–∏. –ï—Å–ª–∏ —É–ø—Ä–æ—Å—Ç–∏—Ç—å, —Ç–æ –≤—ã—Ä—É—á–∫–∞ = –∫–æ–Ω–≤–µ—Ä—Å–∏—è * —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫. –ï—Å–ª–∏ –≤—ã—Ä—É—á–∫–∞ –≤—ã—Ä–æ—Å–ª–∞, —Ç–æ –º–æ–≥ —É–≤–µ–ª–∏—á–∏—Ç—Å—è —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫, –∞ –º–æ–≥–ª–æ —É–≤–µ–ª–∏—á–∏—Ç—Å—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ (—Ç.–µ. –∫–æ–Ω–≤–µ—Ä—Å–∏—è).

–ó–¥–µ—Å—å –Ω—É–∂–Ω–æ –¥–∞—Ç—å –ø–æ–ª–Ω–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ –≤–æ–∑–º–æ–∂–Ω—ã—Ö –ø—Ä–∏—á–∏–Ω —Ç–∞–∫–æ–≥–æ –ø–æ–≤–µ–¥–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞. –ï—Å–ª–∏ –º—ã –ø—Ä–µ–¥–ª–∞–≥–∞–µ–º —Ç–æ–ª—å–∫–æ –æ–¥–Ω—É –ø—Ä–∏—á–∏–Ω—É —Å–∫–∞—á–∫–∞, –º–æ–∂–Ω–æ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–∏—Ç—å, —á—Ç–æ –º—ã –Ω–µ –≤–ø–æ–ª–Ω–µ –ø–æ–Ω–∏–º–∞–µ–º –≤—Å—é –∫–∞—Ä—Ç–∏–Ω—É.

**2.2 –ö—É–º–º—É–ª—è—Ç–∏–≤–Ω—ã–π —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –ø–æ –≥—Ä—É–ø–ø–∞–º.**

```python
cum_check_a = cummulative_table[cummulative_table['group'] == 'A'][['date', 'av_check', 'orders']]
cum_check_b = cummulative_table[cummulative_table['group'] == 'B'][['date', 'av_check', 'orders']]

plt.plot(cum_check_a['date'], cum_check_a['av_check'], label = 'A')
plt.plot(cum_check_b['date'], cum_check_b['av_check'], label = 'B')
plt.ylabel('–í—ã—Ä—É—á–∫–∞, –º–ª–Ω.—Ä—É–±–ª–µ–π')
plt.title('–î–∏–Ω–∞–º–∏–∫–∞ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –ø–æ –≥—Ä—É–ø–ø–∞–º', color = 'Blue')
plt.xticks(rotation=90)
plt.legend();
```

–ò—Å–∫–ª—é—á–∞—è 1,2,13,14 –∞–≤–≥—É—Å—Ç–∞, —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫ –≥—Ä—É–ø–ø—ã –í –≤—ã—à–µ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –≥—Ä—É–ø–ø—ã –ê. –í —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ —Ç–æ–∂–µ –≤–∏–¥–∏–º –≤—Å–ø–ª–µ—Å–∫ –≤ –≥—Ä—É–ø–ø–µ –í 18 –∞–≤–≥—É—Å—Ç–∞. –ü—Ä–µ–¥–ø–æ–ª–æ–∂—É, —á—Ç–æ –≤ —ç—Ç–æ—Ç –¥–µ–Ω—å –±—ã–ª–∞/–∏ —Å–æ–≤–µ—Ä—à–µ–Ω—ã –¥–æ—Ä–æ–≥–∏–µ –ø–æ–∫—É–ø–∫–∏, –ø–æ–≤–ª–∏—è–≤—à–∏–µ –∏ –Ω–∞ —Å—Ä–µ–¥–Ω–∏–π —á–µ–∫, –∏ –Ω–∞ –≤—ã—Ä—É—á–∫—É.

**2.10 –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º**

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º ‚Äî –±–µ–∑ —É–¥–∞–ª–µ–Ω–∏—è –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–î–ª—è —ç—Ç–æ–≥–æ –≤–≤–µ–¥–µ–º 2 –≥–∏–ø–æ—Ç–µ–∑—ã:

–û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (–ù0) - —Ä–∞–∑–ª–∏—á–∏–π –≤ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–µ—Ç.

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (–ù1) - —Ä–∞–∑–ª–∏—á–∏–µ –≤ –∫–æ–Ω–≤–µ—Ä—Å–∏–∏ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å.

```python
visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

visitors_a_cummulative = visitors_a_daily.apply(
    lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_a': 'sum'}
    ),
    axis=1,
)
visitors_a_cummulative.columns = ['date', 'visitors_a_cummulative']

visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

visitors_b_cummulative = visitors_b_daily.apply(
    lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'visitors_per_date_b': 'sum'}
    ),
    axis=1,
)
visitors_b_cummulative.columns = ['date', 'visitors_b_cummulative']

orders_a_daily = (
    orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
orders_a_daily.columns = ['date', 'orders_per_date_a', 'revenue_per_date_a']

orders_a_cummulative = orders_a_daily.apply(
    lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_a': 'sum', 'revenue_per_date_a': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_a_cummulative.columns = [
    'date',
    'orders_cummulative_a',
    'revenue_cummulative_a',
]

orders_b_daily = (
    orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
orders_b_daily.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']

orders_b_cummulative = orders_b_daily.apply(
    lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_b': 'sum', 'revenue_per_date_b': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
orders_b_cummulative.columns = [
    'date',
    'orders_cummulative_b',
    'revenue_cummulative_b',
]

data = (
    orders_a_daily.merge(
        orders_b_daily, left_on='date', right_on='date', how='left'
    )
    .merge(orders_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(orders_b_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_a_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_b_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_b_cummulative, left_on='date', right_on='date', how='left')
)
data.head(10)
#—Å–æ–∑–¥–∞–¥–∏–º –¥–∞—Ç–∞—Å–µ—Ç —Ä–∞—Å—Å—á–µ—Ç–∞–º –Ω–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –ø–æ –≥—Ä—É–ø–ø–∞–º
```

revenue_per_date_a ‚Äî —Å—É–º–º–∞—Ä–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ A;

orders_daily_a ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ B;

revenue_per_date_a ‚Äî —Å—É–º–º–∞—Ä–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ B;

orders_cummulative_a ‚Äî —Å—É–º–º–∞—Ä–Ω–æ–µ —á–∏—Å–ª–æ –∑–∞–∫–∞–∑–æ–≤ –¥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –¥–∞—Ç—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤ –≥—Ä—É–ø–ø–µ A;

revenue_cummulative_a ‚Äî —Å—É–º–º–∞—Ä–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –¥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –¥–∞—Ç—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤ –≥—Ä—É–ø–ø–µ A;

orders_cummulative_b ‚Äî —Å—É–º–º–∞—Ä–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–∫–∞–∑–æ–≤ –¥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –¥–∞—Ç—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤ –≥—Ä—É–ø–ø–µ B;

revenue_cummulative_b ‚Äî —Å—É–º–º–∞—Ä–Ω–∞—è –≤—ã—Ä—É—á–∫–∞ –¥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –¥–∞—Ç—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤ –≥—Ä—É–ø–ø–µ B;

visitors_per_date_a ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ A;

visitors_per_date_b ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –≤ –≤—ã–±—Ä–∞–Ω–Ω—É—é –¥–∞—Ç—É –≤ –≥—Ä—É–ø–ø–µ B;

visitors_cummulative_a ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –¥–∞—Ç—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤ –≥—Ä—É–ø–ø–µ A;

visitors_cummulative_b ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π –¥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–π –¥–∞—Ç—ã –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ –≤ –≥—Ä—É–ø–ø–µ B.

<div class="alert alert-warning">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ ü§ì<a class="tocSkip"> </h2>

–ù—É–∂–Ω–æ –ª–∏ –Ω–∞–º —Å–æ–±–∏—Ä–∞—Ç—å –≤—Å—é —Ç–∞–±–ª–∏—Ü—É, –µ—Å–ª–∏ –º—ã –∏—Å–ø–æ–ª—å–∑—É–µ–º –Ω–µ –≤—Å–µ –ø–æ–∫–∞–∑–∞—Ç–µ–ª–∏ –≤ —Ä–∞—Å—á–µ—Ç–∞—Ö?)

```python
orders_by_users_a = (
    orders[orders['group'] == 'A']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
orders_by_users_a.columns = ['visitor_id', 'orders']

orders_by_users_b = (
    orders[orders['group'] == 'B']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
orders_by_users_b.columns = ['visitor_id', 'orders']
#—Å–æ–∑–¥–∞–µ–º 2 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
```

```python
example_a = pd.concat([orders_by_users_a['orders'],
                      pd.Series(
                          0, 
                          index=np.arange(data['visitors_per_date_a'].sum() - 
                                          len(orders_by_users_a['orders'])), 
                          name='orders')],axis=0
                    )

example_b = pd.concat([orders_by_users_b['orders'],
                      pd.Series(
                          0, 
                          index=np.arange(data['visitors_per_date_b'].sum() - 
                                          len(orders_by_users_b['orders'])), 
                          name='orders')],axis=0
                    )
#–æ–±—ä—è–≤–∏–ª–∏ 2 –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–∏—Ö —Ç–µ—Å—Ç–æ–≤
```

```python
def diff_function(data_a, data_b):
    alpha = 0.05
    p_value = stats.mannwhitneyu(data_a, data_b)[1]
    print("P-value: {0:.3f}".format(p_value))

    if (p_value < alpha):
        print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
    else:
        print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
    
    print("–û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –í –∫ –ê: {0:.3%}".format(data_b.mean() / data_a.mean()-1))
#—Å–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –±—É–¥–µ—Ç –ø—Ä–æ–≤–µ—Ä—è—Ç—å –≥–∏–ø–æ—Ç–µ–∑—ã
```

–ó–∞–¥–∞–¥–∏–º —É—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha = 0.05.
–í—ã–≤–µ–¥–µ–º —Ä–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –í –∏ –ê.

```python
diff_function(example_a, example_b)
#–ø–µ—Ä–µ–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ
```

–ü–æ —Å—ã—Ä—ã–º –¥–∞–Ω–Ω—ã–º —Ä–∞–∑–ª–∏—á–∏–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å (p-value < alpha). –û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É, –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –≤—ã–∏–≥—Ä—ã—à –≥—Ä—É–ø–ø—ã –í = 13,8%.

–ì–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ –∏ –∑–Ω–∞—á–µ–Ω–∏–µ p-value –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞–Ω–æ –ø—Ä–∞–≤–∏–ª—å–Ω–æ. –•–æ—Ä–æ—à–∞—è —Ä–∞–±–æ—Ç–∞.

**2.11 –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º**

–ü–æ—Å—á–∏—Ç–∞–µ–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´—Å—ã—Ä—ã–º¬ª –¥–∞–Ω–Ω—ã–º ‚Äî –±–µ–∑ —É–¥–∞–ª–µ–Ω–∏—è –∞–Ω–æ–º–∞–ª—å–Ω—ã—Ö –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π.

–î–ª—è —ç—Ç–æ–≥–æ –≤–≤–µ–¥–µ–º 2 –≥–∏–ø–æ—Ç–µ–∑—ã:

–û—Å–Ω–æ–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (–ù0) - —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –Ω–µ—Ç.

–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ (–ù1) - —Ä–∞–∑–ª–∏—á–∏–µ –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å.

```python
diff_function(orders[orders['group'] == 'A']['revenue'], orders[orders['group'] == 'B']['revenue'])
```

P-value –±–æ–ª—å—à–µ, —á–µ–º alpha. –ù—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ–º. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–π –ø—Ä–∏—Ä–æ—Å—Ç –í –∫ –ê - 25,8%

<div class="alert alert-warning">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ ü§ì<a class="tocSkip"> </h2>
    
–°—Ç–æ–∏—Ç –¥–æ–ø–æ–ª–Ω–∏—Ç—å, —á—Ç–æ –ø–æ—Å–∫–æ–ª—å–∫—É –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ–µ —Ä–∞–∑–ª–∏—á–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞ –º–µ–∂–¥—É —Å–µ–≥–º–µ–Ω—Ç–∞–º–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç ~26%, –Ω–∞–±–ª—é–¥–∞–µ–º—ã–µ —Ä–∞–∑–ª–∏—á–∏—è –Ω–µ —è–≤–ª—è—é—Ç—Å—è –∑–Ω–∞—á–∏–º—ã–º–∏ –ø—Ä–∏ —Ç–∞–∫–æ–º —É—Ä–æ–≤–Ω–µ –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏, –∞ —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ –±—ã–ª–∏ –ø–æ–ª—É—á–µ–Ω—ã —Å–ª—É—á–∞–π–Ω–æ. </div>

**2.12 –ü–æ—Å—á–∏—Ç–∞—Ç—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫—É—é –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ –∑–∞–∫–∞–∑–æ–≤ –Ω–∞ –ø–æ—Å–µ—Ç–∏—Ç–µ–ª—è –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º.**

```python
users_with_many_orders = pd.concat(
    [
        orders_by_users_a[orders_by_users_a['orders'] > 4]['visitor_id'],
        orders_by_users_b[orders_by_users_b['orders'] > 4]['visitor_id'],
    ],
    axis=0,
)
users_with_expensive_orders = orders[orders['revenue'] > 58233]['visitor_id']
abnormal_users = (
    pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
    .drop_duplicates()
    .sort_values()
)
sample_a_filtered = pd.concat(
    [
        orders_by_users_a[
            np.logical_not(orders_by_users_a['visitor_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sample_b_filtered = pd.concat(
    [
        orders_by_users_b[
            np.logical_not(orders_by_users_b['visitor_id'].isin(abnormal_users))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
#–æ—á–∏—Å—Ç–∏–º –¥–∞–Ω–Ω—ã–µ, –∏—Å—Ö–æ–¥—è –∏–∑ —Ç–µ—Ö –ø—Ä–µ–¥–µ–ª–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã–±—Ä–∞–ª–∏ —Ä–∞–Ω–µ–µ
```

<div class="alert alert-warning">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ ü§ì<a class="tocSkip"> </h2>

–ß—Ç–æ–±—ã –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º –º—ã –º–æ–≥–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∞–Ω–∞–ª–∏–∑ –ø–æ–≤—Ç–æ—Ä–Ω–æ, –±—É–¥–µ—Ç –ª—É—á—à–µ –æ—Ç–∫–∞–∑–∞—Ç—å—Å—è –æ—Ç  —Ä—É—á–Ω–æ–≥–æ –≤–≤–æ–¥–∞ –æ–±–æ–∑–Ω–∞—á–∞–µ–º—ã—Ö –≥—Ä–∞–Ω–∏—Ü, –∏ –∑–∞–º–µ–Ω–∏—Ç—å –Ω–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –∫–æ–¥—É –æ–±–æ–∑–Ω–∞—á–µ–Ω–∏—è (–ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª–∏))

</div>

```python
diff_function(sample_a_filtered, sample_b_filtered)
```

–ù–∞ –æ—á–∏—â–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –µ—Å—Ç—å (p-value < alpha). –ù—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ—Ç–≤–µ—Ä–≥–∞–µ–º. –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ 15,3%.

**2.13 –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –∑–Ω–∞—á–∏–º–æ—Å—Ç—å —Ä–∞–∑–ª–∏—á–∏–π –≤ —Å—Ä–µ–¥–Ω–µ–º —á–µ–∫–µ –∑–∞–∫–∞–∑–∞ –º–µ–∂–¥—É –≥—Ä—É–ø–ø–∞–º–∏ –ø–æ ¬´–æ—á–∏—â–µ–Ω–Ω—ã–º¬ª –¥–∞–Ω–Ω—ã–º.**
